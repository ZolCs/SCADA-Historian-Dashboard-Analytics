```python
# Import libraries

import random
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
```

```python
# Generating random temperature values with timestamp

current_time = datetime(2026, 6, 11, 6, 00) 
base_temperature = 40
temperature=[]
timestamp=[]
    
for i in range(6):
    for j in range(20):
        current_time += timedelta(minutes=1)            
        increment = random.randint(-2, 2)
        base_temperature += increment
        timestamp.append(current_time)
        temperature.append(base_temperature)
                        
    for k in range(10):
        current_time += timedelta(minutes=1) 
        increment_1 = random.randint(2, 8)
        base_temperature += increment_1
        timestamp.append(current_time)
        temperature.append(base_temperature)
            
    for l in range(3):
        current_time += timedelta(minutes=1)
        increment_2 = random.randint(20, 30)
        base_temperature += increment_2
        timestamp.append(current_time)
        temperature.append(base_temperature)
            
    for m in range(8):
        current_time += timedelta(minutes=1)
        increment_3 = random.randint(11, 21)
        base_temperature -= increment_3
        timestamp.append(current_time)
        temperature.append(base_temperature)
            
    for n in range(20):
        current_time += timedelta(minutes=1)
        increment_4 = random.randint(-2, 2)
        base_temperature += increment_4
        timestamp.append(current_time)
        temperature.append(base_temperature)
            
print(len(temperature))
print(temperature)
print(len(timestamp))
print(type(timestamp[0]))
```

```python
# Generating random pressure values

base_pressure = 1.5
pressure=[]
s=0
length=0
length=len(temperature)-10

for o in range(1):
        for p in range(10):
            increment_5 = random.uniform(-0.1,0.1)
            base_pressure += increment_5
            pressure.append(round(base_pressure,2))
        for q in range (length):
            increment_6 = random.uniform(0,1.05)
            increment_7 = random.uniform(0,0.5)
            increment_8 = random.uniform(-0.05,0.1)
            if (temperature[s+1] - temperature[s]) > 3.5:
                base_pressure += increment_6
                pressure.append(round(base_pressure,2))
                s += 1
            elif temperature[s+1] < temperature[s]:
                base_pressure -= increment_7
                pressure.append(round(base_pressure,2))
                s += 1
            else:
                s += 1
                base_pressure += increment_8
                pressure.append(round(base_pressure,2))
                
                    
                
print(pressure)
print(len(pressure))
```

```python
temperature_change=[0]
tc=0
pressure_change=[0]
pc=0
length_2=len(temperature)-1
w=0

for o in range(length_2):
    tc=temperature[w+1] - temperature[w]
    temperature_change.append(tc)
    pc=pressure[w+1] - pressure[w]
    pressure_change.append(round(pc,2))
    w += 1

print(temperature_change)
print(len(temperature_change))
print(pressure_change)
print(len(pressure_change))
```

```python
# Visualization

time = timestamp
temp = temperature
press = pressure

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Time')
ax1.set_ylabel('Temperature (°C)', color=color)
ax1.plot(time, temp, color=color, label="Temperature")
ax1.tick_params(axis='y', labelcolor=color)
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%b-%d %H:%M'))
fig.autofmt_xdate()


ax2 = ax1.twinx() 


color = 'tab:blue'
ax2.set_ylabel('Pressure (bar)', color=color)  
ax2.plot(time, press, color=color, label="Pressure")
ax2.tick_params(axis='y', labelcolor=color)


lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax1.legend(
    lines1 + lines2,
    labels1 + labels2,
    bbox_to_anchor=(0.5, -0.4),
    loc='upper center',
    ncol=2
)

plt.title('Temperature and Pressure Monitoring Dashboard')

fig.tight_layout()  
plt.show()
```

```python
# Send values to a dataframe

df_process_data = pd.DataFrame({"timestamp": timestamp,"temperature": temperature, "pressure": pressure, "temperature_change":temperature_change, "pressure_change": pressure_change})
print(df_process_data)
print(df_process_data.info())
print(df_process_data.describe())
```

```python
# Anomaly detection
```

```python
# Engineering alerts

temperature_rising_limit = 100
temperature_critical_limit = 140
pressure_rising_limit = 4
pressure_critical_limit = 6
s=0

df_engineering_alerts = pd.DataFrame(columns=["timestamp", "variable", "level", "value"])


for index, row in df_process_data.iterrows():
    if row["temperature"] > temperature_rising_limit and row["temperature"] < temperature_critical_limit :
        df_engineering_alerts.loc[s]=[row["timestamp"], "temperature", "warning", row["temperature"]]
        s += 1
        
    if row["temperature"] > temperature_critical_limit:
        df_engineering_alerts.loc[s]=[row["timestamp"], "temperature", "critical", row["temperature"]]
        s += 1

    if row["pressure"] > pressure_rising_limit and row["pressure"] <pressure_critical_limit:
        df_engineering_alerts.loc[s]=[row["timestamp"], "pressure", "warning", row["pressure"]]
        s += 1

    if row["pressure"] > pressure_critical_limit:
        df_engineering_alerts.loc[s]=[row["timestamp"], "pressure", "critical", row["pressure"]]
        s += 1
        
print(df_engineering_alerts)
```

```python
# Statistical alerts

temperature_rising_limit_s = df_process_data["temperature"].quantile(0.85)
temperature_critical_limit_s = df_process_data["temperature"].quantile(0.95)
pressure_rising_limit_s = df_process_data["pressure"].quantile(0.85)
pressure_critical_limit_s = df_process_data["pressure"].quantile(0.95)
s=0

df_statistical_alerts = pd.DataFrame(columns=["timestamp", "variable", "level", "value"])

for index, row in df_process_data.iterrows():
    if row["temperature"] > temperature_rising_limit_s and row["temperature"] < temperature_critical_limit_s :
        df_statistical_alerts.loc[s]=[row["timestamp"], "temperature", "warning", row["temperature"]]
        s += 1
        
    if row["temperature"] > temperature_critical_limit_s:
        df_statistical_alerts.loc[s]=[row["timestamp"], "temperature", "critical", row["temperature"]]
        s += 1

    if row["pressure"] > pressure_rising_limit_s and row["pressure"] <pressure_critical_limit_s:
        df_statistical_alerts.loc[s]=[row["timestamp"], "pressure", "warning", row["pressure"]]
        s += 1

    if row["pressure"] > pressure_critical_limit_s:
        df_statistical_alerts.loc[s]=[row["timestamp"], "pressure", "critical", row["pressure"]]
        s += 1
        
print(df_statistical_alerts)
```

```python
import sqlite3
```

```python
sqliteConnection = sqlite3.connect('SCADA_Historian.db')
cursor = sqliteConnection.cursor()
```

```python
# Create table: Pocess_Data

cursor.execute("DROP TABLE IF EXISTS process_data")

table_creation_query = """
    CREATE TABLE process_data (
        ID INTEGER PRIMARY KEY NOT NULL,
        timestamp TEXT,
        temperature INTEGER,
        pressure REAL,
        temperature_change INTEGER,
        pressure_change REAL
    );
"""
cursor.execute(table_creation_query)
```

```python
# Create table: engineering_alerts

cursor.execute("DROP TABLE IF EXISTS engineering_alerts")

table_creation_query = """
    CREATE TABLE engineering_alerts (
        ID INTEGER PRIMARY KEY NOT NULL,
        timestamp TEXT,
        variable TEXT,
        level TEXT,
        value REAL
    );
"""
cursor.execute(table_creation_query)
```

```python
# Create table: statistical_alerts

cursor.execute("DROP TABLE IF EXISTS statistical_alerts")

table_creation_query = """
    CREATE TABLE statistical_alerts (
        ID INTEGER PRIMARY KEY NOT NULL,
        timestamp TEXT,
        variable TEXT,
        level TEXT,
        value REAL
    );
"""
cursor.execute(table_creation_query)
```
